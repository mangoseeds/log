```python
# 랜덤 데이터 생성 함수 정의
def generate_random_data():
    return {
        '일기': 0,
        '요약': 0,
        '해시태그': 0
    }

# 100개의 행을 갖는 데이터프레임 생성
data = [generate_random_data() for _ in range(500)]
df = pd.DataFrame(data)

# 데이터프레임 일부 확인
print(df.head())
```

```python
idx = 0
```

```python
for t in text.split("[일기 "):
  try:
    diary, summary = t.split("요약")
    summary, tag = summary.split("[해시태그]")
    df.iloc[idx, 0] = text[22:].replace("\n"," ").strip()
    df.iloc[idx, 1] = summary.replace("\n"," ")[2:].strip()
    df.iloc[idx, 2] = tag.replace("\n"," ").strip()
    idx+=1
  except:
    continue
```
